```python
def calculate_value(num_bins, mass, intensity):
    lower_mass_bin = None
    upper_mass_bin = None
    lower_mass_bin_intensity = None
    upper_mass_bin_intensity = None
    multiplier = num_bins/MAX_MASS
    
    if mass == 1:
        lower_mass_bin = 0
        upper_mass_bin = 1
        lower_mass_bin_intensity = intensity
        upper_mass_bin_intensity = 0
        return [(lower_mass_bin, lower_mass_bin_intensity), (upper_mass_bin, upper_mass_bin_intensity)]
    elif mass == 1000:
        lower_mass_bin = num_bins - 2
        upper_mass_bin = num_bins - 1
        lower_mass_bin_intensity = 0
        upper_mass_bin_intensity = intensity
        return [(lower_mass_bin, lower_mass_bin_intensity), (upper_mass_bin, upper_mass_bin_intensity)]
    elif mass in range(2, 1000):
        lower_mass_bin = int(mass * multiplier) - 1
        upper_mass_bin = int(mass * multiplier) + 1
        lower_mass_bin_intensity = 0
        upper_mass_bin_intensity = 0
        return [(lower_mass_bin, lower_mass_bin_intensity), (int(mass), intensity), (upper_mass_bin, upper_mass_bin_intensity)]

    if mass > 1 and mass not in range(2, 1000):
        lower_mass_bin = math.floor(float(mass) * multiplier) - 1
    if mass < 1000 and mass not in range (2, 1000):
        upper_mass_bin = math.ceil(float(mass) * multiplier) - 1

    if upper_mass_bin is not None:
        lower_mass_bin_intensity = (-(mass * multiplier - 1) + upper_mass_bin) * intensity
    if lower_mass_bin is not None:
        upper_mass_bin_intensity = ((mass * multiplier - 1) - lower_mass_bin) * intensity

    return [(lower_mass_bin, lower_mass_bin_intensity), (upper_mass_bin, upper_mass_bin_intensity)]
```

```python
def write_binned_files(num_bins=1000):
    binned_datapath = "G:\\Dev\\Data\\{}\\GNPS Python Binned".format(num_bins)
    
    for file in os.listdir(datapath):
        filepath = os.path.join(datapath, file)
        binned_values = np.zeros(num_bins, dtype=float)
        with open(filepath, 'r') as f:
            filename = f.name
            unsplit_lines = list(islice(f, 9, None))
            for line in unsplit_lines:
                if ' ' in line:  # Only lines with mass and intensity values have a space. Ignores label/blank lines
                    split_line = line.split()
                    mass = float(split_line[0])
                    if mass <= MAX_MASS: # If fragment isn't too big
                        rounded_mass = round(mass * (num_bins/MAX_MASS))
                        mass_bin = int((mass * (num_bins/MAX_MASS))) - 1  # Bin fragment belongs in.
                        intensity = float(split_line[1])
                        binned_values[mass_bin] = binned_values[mass_bin] + intensity  # Sum intensities for bin
                        
        binned_filename = file.split(".")[0] + " Binned.txt"
        binned_filepath = os.path.join(binned_datapath, binned_filename)

        with open(binned_filepath, 'w') as f:  # Write bins and intensities to new file.
            for index, intensity in enumerate(binned_values):
                f.write(str(index+1) + "  " + str(intensity) + "\n")


def write_binned_files_triangle(num_bins=1000):
    binned_datapath = "G:\\Dev\\Data\\{}_Triangle\\GNPS Python Binned".format(num_bins)
    
    for file in os.listdir(datapath):
        filepath = os.path.join(datapath, file)
        binned_values = np.zeros(num_bins, dtype=float)
        with open(filepath, 'r') as f:
            filename = f.name
            unsplit_lines = list(islice(f, 9, None))
            for line in unsplit_lines:
                if ' ' in line:  # Only lines with mass and intensity values have a space. Ignores label/blank lines
                    split_line = line.split()
                    mass = float(split_line[0]) # Mass of fragment, to nearest Da
                    if mass <= MAX_MASS: # If fragment isn't too big
                        intensity = float(split_line[1])
                        for index, (mass_bin, intensity) in enumerate(calculate_value(num_bins, mass, intensity)):
                            binned_values[mass_bin] = binned_values[mass_bin] + intensity  # Sum intensities for bin
        binned_filename = file.split(".")[0] + " Binned.txt"
        binned_filepath = os.path.join(binned_datapath, binned_filename)
        
        with open(binned_filepath, 'w') as f:  # Write bins and intensities to new file.
            for index, intensity in enumerate(binned_values):
                f.write(str(index+1) + "  " + str(intensity) + "\n")
```

```python
print("Binning 1000")
write_binned_files(1000)

print("Binning 10000")
write_binned_files(10000)

print("Binning 1000 triangle")
write_binned_files_triangle(1000)

print("Binning 10000 triangle")
write_binned_files_triangle(10000)
```

```python
# Load fragment data for IDing
dtype = [('fragment', 'U25'), ('mass', float)]
fragments = np.loadtxt(mass_path, dtype=dtype, delimiter=',')

def normalize_and_filter(spectrum, max_value=1000, min_percent=0.005):
    min_value = max_value * min_percent
    # print(min_value)
    max_intensity = np.amax(spectrum)  # Find max intensity
    spectrum = spectrum / max_intensity  # Normalize to 0-1
    spectrum = spectrum * max_value  # Scale all values
    filtered = np.where(spectrum < min_value, 0, spectrum)  # Set values below threshold to 0.
    return filtered


def top_six_filter(spectrum, num_bins):
    filtered_spectrum = np.zeros(spectrum.shape, float)
    for i in range(len(spectrum)):  # For each mass bin
        low_end = 0
        if i < 50 * (num_bins//MAX_MASS):
            low_end = i
        if i >= 50 * (num_bins//MAX_MASS):
            low_end = 50 * (num_bins//MAX_MASS)
        
        window_comparison = np.less(spectrum[i], spectrum[i-low_end:(i+50 * (num_bins//MAX_MASS))])
        if np.sum(window_comparison) < 7:  # If value is among top 6 in 1000Da range, add it to filtered array.
            filtered_spectrum[i] = spectrum[i]
            
    return filtered_spectrum


def scan_for_fragments(spectrum, num_bins):
    intensity_only = spectrum[:,1]
    for i, fragment in enumerate(fragments):
        f_mass = float(fragment['mass'])
        f_mass_index = int(round(f_mass * (num_bins/MAX_MASS)))
        
        # Slice intensities twice so bins to be compared share index.
        spec = intensity_only[:(intensity_only.size - f_mass_index)]
        comp_spec = intensity_only[f_mass_index:]
        pairs = np.where(np.logical_and(spec > 0, comp_spec > 0))  # Compare bins separated by residue mass
        f_intensity = 0  # Default value for fragment presence
        if pairs[0].size > 0:
            # Sum up intensities of fragments separated by fragment mass and take highest.
            f_intensity = np.amax(spec[pairs]*comp_spec[pairs])
            #if f_intensity > 1000:
                #f_intensity = 1000
        # Append fragment information to spectrum
        new_row = np.array([intensity_only.size+i+1, f_intensity], dtype=float)
        spectrum = np.vstack([spectrum, new_row])
    return spectrum


def process_files(binned_datapath, filtered_datapath, double_filtered_datapath, fragment_id_datapath, num_bins):
    for file in os.listdir(binned_datapath):
        print(file)
        filepath = os.path.join(binned_datapath, file)
        data = np.loadtxt(filepath, np.float32)
        # Scale and filter all below threshold. Column 1 contains intensities.
        data[:, 1] = normalize_and_filter(data[:, 1])
        # Save to new file.
        filtered_filename = file.split("Binned")[0] + "Filtered.txt"
        filtered_filepath = os.path.join(filtered_datapath, filtered_filename)
        np.savetxt(filtered_filepath, data, fmt="%d %f")

        # Top 6 Filter
        data[:, 1] = top_six_filter(data[:, 1], num_bins)

        double_filtered_filename = file.split("Binned")[0] + "Filtered Twice.txt"
        double_filtered_filepath = os.path.join(double_filtered_datapath, double_filtered_filename)
        np.savetxt(double_filtered_filepath, data, fmt="%d %f")

        # Identify potential fragments
        data = scan_for_fragments(data, num_bins)

        fragment_id_filename = file.split("Binned")[0] + " With Fragments.txt"
        fragment_id_filepath = os.path.join(fragment_id_datapath, fragment_id_filename)
        np.savetxt(fragment_id_filepath, data, fmt="%d %f")
```

```python
print("Processing 1000")
process_files(binned_datapath_1000, filtered_datapath_1000, double_filtered_datapath_1000, fragment_id_datapath_1000, 1000)

print("Processing 10000")
process_files(binned_datapath_10000, filtered_datapath_10000, double_filtered_datapath_10000, fragment_id_datapath_10000, 10000)

print("Processing 1000 triangle")
process_files(binned_datapath_1000_triangle, filtered_datapath_1000_triangle, double_filtered_datapath_1000_triangle, fragment_id_datapath_1000_triangle, 1000)

print("Processing 10000 triangle")
process_files(binned_datapath_10000_triangle, filtered_datapath_10000_triangle, double_filtered_datapath_10000_triangle, fragment_id_datapath_10000_triangle, 10000)
```

```python
def save_to_master(input_path, output_path):    
    all_files = os.listdir(input_path)

    with open(output_path, 'w') as f:  # The master file to write to
        for file in all_files:  # For each filtered file
            file_path = os.path.join(input_path, file)
            with open(file_path, 'r') as d:  # Open and read the file
                print(file)
                mol_id = file.split()[0]
                lines = list(islice(d, 0, None))
                for line in lines:  # For each line (feature) in the file
                    split_line = line.split()
                    mass = split_line[0]
                    intensity = float(split_line[1])
                    if intensity != 0.0:  # Write to master file only if intensity is non-zero
                        f.write(mol_id + " ")  # Molecule ID
                        f.write(mass + " ")  # Mass bin
                        f.write(str(intensity) + "\n")  # Intensity for mass bin
```

```python
# Save to 1000 master
print("Saving 1000")
save_to_master(double_filtered_datapath_1000, final_path_1000)
save_to_master(fragment_id_datapath_1000, final_fragment_path_1000)

# Save to 10000 master
print("Saving 10000")
save_to_master(double_filtered_datapath_10000, final_path_10000)
save_to_master(fragment_id_datapath_10000, final_fragment_path_10000)

# Save to 1000_triangle master
print("Saving 1000 triangle")
save_to_master(double_filtered_datapath_1000_triangle, final_path_1000_triangle)
save_to_master(fragment_id_datapath_1000_triangle, final_fragment_path_1000_triangle)

# Save to 10000_triangle master
print("Saving 10000 triangle")
save_to_master(double_filtered_datapath_10000_triangle, final_path_10000_triangle)
save_to_master(fragment_id_datapath_10000_triangle, final_fragment_path_10000_triangle)
```
